    Things to keep in mind:
        • The input for the function should be the given dataframe
        • The input represents the sales records. The columns are 'Date', 'Product', 'Quantity', and 'Price'.
        • The output format is expected to be a dataframe with the product names and their total sales.

```python
import pandas as pd

# Define a function to calculate total sales based on quantity and price.
def calculate_total_sales(df):
    # Check if 'Price' and 'Quantity' columns are present in the DataFrame.
    if 'Price' in df.columns and 'Quantity' in df.columns:
        # Calculate 'Total Sales' by multiplying 'Quantity' and 'Price' columns.
        df['Total Sales'] = df['Quantity'] * df['Price']
        
        # Group the DataFrame by 'Product' and sum the 'Total Sales' for each product.
        total_sales_df = df.groupby('Product')['Total Sales'].sum().reset_index()
        
        # Return the resulting DataFrame with 'Product' and 'Total Sales'.
        return total_sales_df
    else:
        # If 'Price' or 'Quantity' columns are not present, print an error message.
        print("Required columns 'Price' and 'Quantity' not found in the DataFrame.")
        # Return None to indicate that the calculation couldn't be performed.
        return None

# Sample data for sales with columns: Date, Product, Quantity, and Price.
data = {'Date': ['2023-09-01', '2023-09-01', '2023-09-02', '2023-09-02'],
        'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop'],
        'Quantity': [2, 5, 3, 1],
        'Price': [1000, 20, 50, 100]}

# Create a DataFrame using the sample data.
sales_data = pd.DataFrame(data)

# Call the calculate_total_sales function with the sales_data DataFrame.
total_sales_result = calculate_total_sales(sales_data)

# Print the resulting DataFrame showing 'Product' and corresponding 'Total Sales'.
print(total_sales_result)
```

### 3) Reverse a string with constraints
    Constraints:
        • You can reverse the letters
        • You cannot reverse the symbols or the numbers and they should retain their index positions in the output from the
        input.
        
    Input:
        'a6v4j7k%pt*'
    Output:
        't6p4k7j%va*'

```python
def reverse_string_with_constraints(input_string):
    # Separate letters, symbols, and numbers
    letters = [char for char in input_string if char.isalpha()]
    symbols_and_numbers = [char for char in input_string if not char.isalpha()]
    
    # Reverse the order of letters
    reversed_letters = letters[::-1]
    
    # Initialize an empty string to store the final output
    output_string = ''
    
    # Initialize an index for the reversed_letters list
    letter_index = 0
    
    # Iterate through each character in the input string
    for char in input_string:
        # If the character is a letter, append the reversed letter to the output
        if char.isalpha():
            output_string += reversed_letters[letter_index]
            letter_index += 1
        else:
            # If the character is not a letter, append it directly to the output
            output_string += char

    # Return the final output string
    return output_string

# Example usage
input_str = 'a6v4j7k%pt*'
output_str = reverse_string_with_constraints(input_str)
print(output_str)
```

### 4) Create a function that prints a pattern of 1's and 0's based on the given requirements
    Requirements:
        • The function should only take one user given input, which is the size of the matrix
        • When the size is less than 3, give the option to ask for input again. If the given input is wrong then
        print the text 'Size can be greater than 3 only, try again later'
        • When the given user input is odd, then print the pattern X using 1's in the pattern of 0's and 1's
        • When the given user input is even, then print the pattern Z using 0's in the pattern of 0's and 1's
        • The output doesn't strictly have to be in the given format (Can be an array, list of lists, etc)

        An example for size 3:
        1 0 1
        0 1 0
        1 0 1
        
    Please mention if any other assumptions were required/made

```python
def print_pattern():
    while True:
        try:
            # Get the size of the matrix from the user
            size = int(input("Enter the size of the matrix: "))
            
            # Check if the size is less than 3
            if size < 3:
                print("Size can be greater than 3 only, try again later")
            else:
                # If the size is valid, exit the loop
                break
        except ValueError:
            # Handle the case where the user enters a non-integer value
            print("Invalid input. Please enter a valid integer.")

    # Initialize an empty list to store the pattern
    pattern = []

    # Iterate through rows
    for i in range(size):
        row = []
        # Iterate through columns
        for j in range(size):
            # Check the conditions for filling the matrix based on size
            if size % 2 == 1:  # Odd-sized matrix
                if i == j or i + j == size - 1:
                    row.append(1)
                else:
                    row.append(0)
            else:  # Even-sized matrix
                if i == 0 or i == size - 1 or j == size - i - 1:
                    row.append(0)
                else:
                    row.append(1)
        # Append the row to the pattern
        pattern.append(row)

    # Print the pattern
    for row in pattern:
        print(' '.join(map(str, row)))

# Call the function to execute the pattern printing
print_pattern()
```
